```python
import modelo_update_sasaki
import os
```

```python

ruta_base = "/home/dorito/onedrive_coocentral/Reportes Qfield/0412/"

carpetas_validas = []

for nombre_carpeta in os.listdir(ruta_base):
    ruta_carpeta = os.path.join(ruta_base, nombre_carpeta)

    # Solo queremos carpetas del nivel principal
    if not os.path.isdir(ruta_carpeta):
        continue

    # Buscar archivos dentro de esa carpeta que empiecen por qfield
    contiene_qfield = any(
        archivo.startswith("Qfield")
        for archivo in os.listdir(ruta_carpeta)
        if os.path.isfile(os.path.join(ruta_carpeta, archivo))
    )

    if contiene_qfield:
        carpetas_validas.append(ruta_carpeta)

print("Corriendo modelo*: ")
for c in carpetas_validas:
    print(c)
    modelo_update_sasaki.execute_model(c,"https://pruebas.local/")
```
